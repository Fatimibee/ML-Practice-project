```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Student_Performance.csv")
data.info()
```

```python
x=data[['Hours Studied','Previous Scores','Sleep Hours','Sample Question Papers Practiced']]
x
```

```python
x
```

```python
y=data[['Performance Index']]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=45)
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predicted=model.predict(x_test)
y_predicted
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_predicted)   
score 

#0.98 tells  model is explaining 98.8% of the variance in the data, which is very good
```

```python
import matplotlib.pyplot as plt

error = y_predicted - y_test

plt.figure(figsize=(6, 4))
plt.hist(error, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Distribution of Prediction Errors")
plt.xlabel("Error")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
import joblib

# Save the model
joblib.dump(model, 'student_performance.pkl')
```
